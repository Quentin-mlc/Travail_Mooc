```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\franc\Downloads\Data\usages.effec1.csv", encoding='latin1')
data2 = pd.read_csv(r"C:\Users\franc\Downloads\Data\effec1.quest.compil.csv",encoding='latin1')
```

```python
data3=pd.merge(data,data2,on="Student_ID")
pd.set_option("display.max_columns", None)
```

```python
data = [["Country_HDI.fin"]]
```

```python
data3["n.vid"] = data3.iloc[:,17:43].sum(axis=1)
```

```python
Means = data3.groupby(['Gender'])["n.vid"].mean()
print(Means)
```

```python
STD = data3.groupby(['Gender'])["n.vid"].std()
print(STD)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Création du barplot avec barres d'erreur
plt.figure(figsize=(8, 5))
sns.barplot(x=Means.index, y=Means.values, capsize=0.2, palette="viridis")

# Ajout des labels et titre
plt.xlabel("IDH")
plt.ylabel("Moyenne de n.vid")
plt.title("Moyenne de n.vid par Genre avec barres d'erreur (écart-type)")

# Ajout d'une grille pour la lisibilité
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Affichage du graphique
plt.show()


```
